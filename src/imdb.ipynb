# IMDB REVIEW COLLECTION

This code consists of a crawler for IMDB that, given the reviews page of a specific movie, it loads the "more reviews" button 10 times and collects all the reviews, labelled with their title.

This notebook doesn't work on collab because of the need to download the CromeDriver tool, which is needed to interact with the webpage when clicking the "load more" from the code.

```python
#import requests to download the page and bs4 to parse
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
from urllib.parse import urljoin


options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver =webdriver.Chrome()
```

```python
import time

driver.get("https://m.imdb.com/title/tt6723592/reviews?ref_=m_tt_urv")
more_buttons = driver.find_elements_by_class_name("ipl-load-more__button")
for x in range(5):
      if more_buttons[0].is_displayed():
          driver.execute_script("arguments[0].click();", more_buttons[0])
          time.sleep(2)
page_source = driver.page_source
```

```python
url = driver.current_url
res = requests.get(url)
soup = BeautifulSoup(page_source,"lxml")

titles = []
reviews =[]

for item in soup.select(".review-container"):
    title = item.select(".title")[0].text
    review = item.select(".text")[0].text
    #print("Title: {}\n\nReview: {}\n\n".format(title,review))
    titles.append(title)
    reviews.append(review)
```

```python
# dictionary of lists  
dict = {'title': titles, 'review': reviews}  
    
df = pd.DataFrame(dict) 
    
df.head(10)
```
