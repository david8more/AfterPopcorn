# Metrics of the retrieved data

In this notebook we are going to evaluate and analyze all the data that we have been collecting individually. Film by film, we have combined the statistics of the three platforms (YouTube, Reddit and IMDB), and we have built dataframes containing all the necessary information. The metrics that we have found interesting, taking into account the sentiment, are: 



1.   Mean and variance of the compound value given by VADER per platform: This will tell us how positive, negative or neutral is the film reception in each platform, and might help to find patterns.
2.   Number of positive, negative and neutral comments per platform. We dislpay this as a bar chart.
3.   Proportion of positive, negative and neutral words per comment, taking into account the means of each platform. For example, if the mean of positive words is 3.5, and the mean of number of words per comment is 14.5, the proportion is 3.5/14.5. We display this as a pie chart to see clearly the differences.


**Note:** Some plots don't need to be commented with text below them due to the fact that there will have been explained in previous films analysis, and they present really similar behaviours, wo it would be a waste of time to repeat the same. The general concluslions are added at the end of this notebook, and we will cover all the films.

## Avengers: Endgame 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_endgame = pd.read_csv('/PATH/TO/file.csv')
df_endgame.set_index('platrofm', inplace=True, drop=True)
df_endgame = df_endgame.fillna(0) 
df_endgame.head(7)
```

Taking into account the mean compound value of each chapter, we plot the mean and also the variance the values of all the retrieved tweets per episode. The variance is important, as could show up discrepancy in Twitter opinions. 

```python
labels = ['YouTube','Reddit','IMDB']

fig = plt.figure(figsize = (10,7))
y_endgame = df_endgame['compound_mean'].tolist()
var = df_endgame['compound_var'].tolist()
x = np.arange(0,3)
var_array = np.array(var)
error = var_array/2
plt.plot(x, y_endgame, label='Mean')
plt.fill_between(x, y_endgame-error, y_endgame+error, color='#ff7f0e', alpha=0.25, label='Variance')
plt.xlabel('Platform')
# plt.xticklabels(labels)
plt.ylabel('Mean & Variance')
plt.title('Endgame Mean and Variance')
plt.grid(linestyle='dotted', linewidth='0.5', color='grey')
plt.axis([0, 2, -0.5, 1])
plt.xticks([0,1,2], labels)
plt.legend()
plt.show()
```

In the figure above, we can observe that IMDB comments are considerably more negative (or in other words, have a lower compound mean). This might be cased by the fact that in IMDB there are more professional people than on the other two platforms, perhaps critics who are dedicated to it. At this stage, we cannot draw any conclusions, except to say that Avenger:Endgame had a worse reception among the professionals than the fandom.



Now we analyze the number of comments per each "emotion" in each chapter.

```python
labels = ['YouTube','Reddit','IMDB']
x = np.arange(len(labels))
num_tweets = df_endgame['num_comments'].to_numpy()
# pos_tweets is the real number
pos_tweets = df_endgame['num_pos_com'].to_numpy()
# pos_prop is the proportion (pos_tweets/all_tweets)
pos_prop = [i / j for i, j in zip(pos_tweets, num_tweets)]

neg_tweets = df_endgame['num_neg_com'].to_numpy()
neg_prop = [i / j for i, j in zip(neg_tweets, num_tweets)]

neu_tweets = df_endgame['num_neu_com'].to_numpy()
neu_prop = [i / j for i, j in zip(neu_tweets, num_tweets)]

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(15)
width = 0.15
pos_bar = ax.bar(x - width*1.15, pos_prop, width, color='#72D26F', label='Positive ')
neg_bar = ax.bar(x ,  neg_prop,  width, color='#DD4D4D', label='Negative ')
neu_bar = ax.bar(x + width*1.15,  neu_prop,  width, color='#FFAE42', label='Neutral ')
ax.set_ylabel('Proportion of comments')
ax.set_xlabel('Platform')
ax.set_title('Endgame Comments analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.grid(linestyle='dotted', linewidth='0.5', color='grey')

ax.legend(loc='upper right')

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(float(round(height,2))),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(pos_bar)
autolabel(neg_bar)
autolabel(neu_bar)
```

In this case, IMDB has a completely opposite opinion on this film, inverting the number of positive and negative comments/reviews. It's remarkable, and we'll see this along all the IMDB metrics, that the neutrality in this platform is a strange phonomena. This could be given by the amount of words used in each comment, much higher than in YouTube or Reddit. 

```python
labels = ['YouTube','Reddit','IMDB']
labels2 = ['Positive','Negative','Neutral']
num_words_per_comment = df_endgame['words_per_comment mean']

pos_words = df_endgame['pos_words/coms mean'] 
pos_proportion = [i / j for i, j in zip(pos_words, num_words_per_comment)]

neg_words = df_endgame['neg_words/coms mean'] 
neg_proportion = [i / j for i, j in zip(neg_words, num_words_per_comment)]

neu_proportion = [(1-(i + j)) for i, j in zip(pos_proportion, neg_proportion)]

sizes1 = [pos_proportion[0], neg_proportion[0], neu_proportion[0]] #Youtube
sizes2 = [pos_proportion[1], neg_proportion[1], neu_proportion[1]] #Reddit
sizes3 = [pos_proportion[2], neg_proportion[2], neu_proportion[2]] #IMDB

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(9)
fig.set_figwidth(21)
fig.suptitle('Avengers:Endgame positive, negative and neutral words per comment', fontsize=24)
explode = (0, 0, 0.1)  
colors = ['#37C959','#EF5B5B','#bfbfbf']
hole_size = 0.5

ax1.pie(sizes1, explode=explode, labels=labels2, colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax1.axis('equal')# Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_xlabel('Youtube',fontsize=20)


ax2.pie(sizes2, explode=explode, labels=labels2,  colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax2.axis('equal')  
ax2.set_xlabel('Reddit',fontsize=20)

ax3.pie(sizes3, explode=explode, labels=labels2,  colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax3.axis('equal')  
ax3.set_xlabel('IMDB',fontsize=20)
```

From these pie charts we could observe that, at least for this film, the most balanced platform regarding the number of positive and negative words per comment/review is IMDB. Nevertheless, as we have computed mean number of words, and mean positive and negative words per comment, there exist the possibility that we are generalizing too much. In this case, this contradicts the above bar chart, as IMDB is the one with more contrast in number of positive and negative reviews.

Also it is important to understand that it is normal to only have the 10% of the words polarized, as in a text, there might be a lot of words such as conjuctions, prepositions, or even names (table) that are neutral, and does not provide anything in that sense, at leas working with unigrams.

```python
labels = ['YouTube','Reddit','IMDB']

fig = plt.figure(figsize = (10,7))
y_martian = df_martian['compound_mean'].tolist()
var = df_martian['compound_var'].tolist()
x = np.arange(0,3)
var_array = np.array(var)
error = var_array/2
plt.plot(x, y_martian, label='Mean')
plt.fill_between(x, y_martian-error, y_martian+error, color='#ff7f0e', alpha=0.25, label='Variance')
plt.xlabel('Platform')
plt.ylabel('Mean & Variance')
plt.title('The Martian Mean and Variance')
plt.grid(linestyle='dotted', linewidth='0.5', color='grey')
plt.axis([0, 2, -0.5, 1])
plt.xticks([0,1,2], labels)
plt.legend()
plt.show()
```

We found necessary to comment this film in particular, because is the one with a strongest change between YouTube/Reddit and IMDB. This film was surprisingly the best received by IMDB reviews. 

Now we analyze the number of tweets per each "emotion" in each chapter.

```python
labels = ['YouTube','Reddit','IMDB']
x = np.arange(len(labels))
num_tweets = df_martian['num_comments'].to_numpy()
# pos_tweets is the real number
pos_tweets = df_martian['num_pos_com'].to_numpy()
# pos_prop is the proportion (pos_tweets/all_tweets)
pos_prop = [i / j for i, j in zip(pos_tweets, num_tweets)]

neg_tweets = df_martian['num_neg_com'].to_numpy()
neg_prop = [i / j for i, j in zip(neg_tweets, num_tweets)]

neu_tweets = df_martian['num_neu_com'].to_numpy()
neu_prop = [i / j for i, j in zip(neu_tweets, num_tweets)]

fig, ax = plt.subplots()
fig.set_figheight(8)
fig.set_figwidth(15)
width = 0.15
pos_bar = ax.bar(x - width*1.15, pos_prop, width, color='#72D26F',label=' Positive ')
neg_bar = ax.bar(x ,  neg_prop,  width, color='#DD4D4D', label=' Negative ')
neu_bar = ax.bar(x + width*1.15,  neu_prop,  width, color='#FFAE42', label=' Neutral ')
ax.set_ylabel('Proportion of comments')
ax.set_xlabel('Platform')
ax.set_title('The Martian Comments analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.grid(linestyle='dotted', linewidth='0.5', color='grey')

ax.legend(loc='upper right')

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(float(round(height,2))),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(pos_bar)
autolabel(neg_bar)
autolabel(neu_bar)
```

```python
labels = ['YouTube','Reddit','IMDB']
labels2 = ['Positive','Negative','Neutral']
num_words_per_comment = df_martian['words_per_comment mean']

pos_words = df_martian['pos_words/coms mean'] 
pos_proportion = [i / j for i, j in zip(pos_words, num_words_per_comment)]

neg_words = df_martian['neg_words/coms mean'] 
neg_proportion = [i / j for i, j in zip(neg_words, num_words_per_comment)]

neu_proportion = [(1-(i + j)) for i, j in zip(pos_proportion, neg_proportion)]

sizes1 = [pos_proportion[0], neg_proportion[0], neu_proportion[0]] #Youtube
sizes2 = [pos_proportion[1], neg_proportion[1], neu_proportion[1]] #Reddit
sizes3 = [pos_proportion[2], neg_proportion[2], neu_proportion[2]] #IMDB

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(9)
fig.set_figwidth(21)
fig.suptitle('The Martian positive, negative and neutral words per comment', fontsize=24)
explode = (0, 0, 0.1)  
colors = ['#37C959','#EF5B5B','#bfbfbf']
hole_size = 0.5

ax1.pie(sizes1, explode=explode, labels=labels2, colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax1.axis('equal')# Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_xlabel('Youtube',fontsize=20)


ax2.pie(sizes2, explode=explode, labels=labels2,  colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax2.axis('equal')  
ax2.set_xlabel('Reddit',fontsize=20)

ax3.pie(sizes3, explode=explode, labels=labels2,  colors=colors, autopct='%1.1f%%',
        shadow=False, startangle=90, wedgeprops=dict(width=hole_size, edgecolor='w'))
ax3.axis('equal')  
ax3.set_xlabel('IMDB',fontsize=20)
```

# Conclusions

Once we have completed the whole analysis for every single film, and computed also some interesting plots for all of them together, we can extract some conclusions: 



*   All the analysis follow a similar pattern: YouTube and Reddit, more social and less profesionalized platforms, present similar results, with slight differences, but the analysis remains quite aligned. On the other hand, IMDB presents more radical results. This platform reviews are much more polarized and thus have a greater variance. 
*   Trying to compare the results obtained for IMDB, we checked the IMDB score for this films. This is not so relevant because IMDb indicates that submitted ratings are filtered and weighted in various ways in order to produce a weighted mean that is displayed for each film, series, and so on. It states that filters are used to avoid ballot stuffing; the method is not described in detail to avoid attempts to circumvent it. In fact, it sometimes produces an extreme difference between the weighted average and the arithmetic mean. But, despite this opacity, we obtained the following: 

>  
* The Martian: 8
* BladeRuner 2049: 8
* The irishman 2049: 7.9
* Once upon a time in Hollywood 2049: 7.6
* Tenet 2049: 7.5
* Endgame 2049: 8.4
* Dunkirk 2049: 7.8

*  As can be seen, the grades do not correspond directly to what we have extracted. To focus on what could have happened to some films, Avengers: Endgame has the best score regarding official IMDB score, and the worst regarding our study. This could be due to the fact that this film has very sad moments (we don't want to make spoilers), like deaths or tough defeats. If this is commented (we are sure this is commented in IMDB reviews because is a central point of the movie), this words could have biased the sentiment analysis. The same could happen to Dunkirk, a war movie.


*  These results may reflect the difference in point of view of people who are familiar with filmmaking and know and follow specific filmmakers and people who simply give their opinion without any previous knowledge, in a purely superficial and subjective basis. Some movies like Once Upon a Time in Hollywood and Tenet could be better considered in IMDB because their filmmakers (Tarantino and Nolan) have a relevant fan base.
